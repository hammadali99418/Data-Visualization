## About this Dataset:
##### `The data was collected from the following` [link](https://www.kaggle.com/code/maryamsikander/instagram-top-users-analysis/comments):
The Top Instagram Accounts Dataset is a collection of 200 rows of data that provides valuable insights into the most popular Instagram accounts across different categories. The dataset contains several columns that provide comprehensive information on each account's performance, engagement rate, and audience size.

1. The "rank": column lists the accounts in order of their popularity on Instagram, starting from the most followed account.

2. The "name": column displays the Instagram handle of the account, which can be used to locate and follow the account on Instagram.

3. The "channel_info": column provides a brief description of the account, such as the type of content it features or the products and services it offers.

4. The "Category": column categorizes the account based on its primary theme or subject matter, such as fashion, sports, entertainment, or food.

5. The "posts": column displays the total number of posts on the account. This column helps to understand the account's level of activity and the amount of content it has produced over time.

6. The "followers": column indicates the number of people who follow the account on Instagram.

7. The "avg likes": column displays the average number of likes that the account's posts receive per post.

8. The "eng rate": column calculates the account's engagement rate by dividing the total number of likes and comments received by the total number of followers, expressed as a percentage.

# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read the dataset

```python
df = pd.read_csv("data.csv")
```

```python
# Max column printing option
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head(5)
```

## Now we visualize the data to create different plots from pandas librray

#### We convert the FOLLOWERS to numeric handling 'M' and 'K' with 'f'

```python
# Convert 'Followers' from string to numeric, handling 'M' and 'K'
def convert_followers(f):
    if 'M' in f:
        return float(f.replace('M', '').replace(',', '').strip()) * 1e6
    elif 'K' in f:
        return float(f.replace('K', '').replace(',', '').strip()) * 1e3
    return float(f)
```

```python
df['Followers'] = df['Followers'].apply(convert_followers)
```

#### The avg.like Convert in to the 'K' in to '  '

```python
# Convert 'Avg. Likes' similarly
df['Avg. Likes'] = df['Avg. Likes'].str.replace('K', '').astype(float) * 1e3
```

```python
df.head(5)
```

# All plots are created by pandas library

# Line Plot

```python
df.plot.line(x='rank', y='Followers', title='Followers by Rank')
plt.ylabel('Followers')
```

This line plot shows how many followers each person has based on their rank. It helps us see how follower counts change with rank.

# Bar Plot

```python
df.plot.bar(x='name', y='Avg. Likes', figsize=(10,5))
plt.xlabel('Influencer Name')
plt.ylabel('Average Likes')
```

compares the average likes received by different people. It clearly shows who gets more likes on their posts.

# Histogram

```python
df['Followers'].plot.hist(bins=20, edgecolor='red')
plt.xlabel('Followers')
plt.ylabel('Frequency')
```

how followers are spread among different people. It tells us how many people fall into specific follower groups.

#  Area Plot

```python
df.set_index('name')['Followers'].plot.area(title='Followers Area Plot', alpha=0.5)
plt.xlabel('Influencer Name')
plt.ylabel('Followers')
```

total number of followers across different people. It gives a visual idea of how follower counts add up.

# Scatter Plot

```python
df.plot.scatter(x='Followers', y='Avg. Likes', title='Followers vs. Average Likes')
plt.xlabel('Followers')
plt.ylabel('Average Likes')
plt.show()
```

The number of followers relates to the average likes on posts. It helps us see if there's a pattern between these two things.

# Pie Plot

```python
df['Category'].value_counts().plot.pie(title='Influencer Categories', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

The different categories and their proportions. It illustrates how much each category makes up of the total, using percentages for easy comparison.

# Hexbin Plot

```python
plt.hexbin(df['Followers'], df['Eng Rate'].str.replace('%', '').astype(float), gridsize=50, cmap='Blues')
plt.title('Hexbin of Followers vs. Engagement Rate')
plt.xlabel('Followers')
plt.ylabel('Engagement Rate')
plt.colorbar(label='Count')
plt.show()
```

 Relationship between the number of followers and the engagement rate. It uses colored hexagons to represent how many points fall into each area, helping to identify patterns in the data.
